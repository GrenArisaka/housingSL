```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler().fit(X_train[:,:8])
def preprocessor(X):
  A = np.copy(X)
  A[:, :8] = scaler.transform(A[:,:8])
  return A

X_train_preprocessed = preprocessor(X_train)
X_train, X_val, X_test = preprocessor(X_train),  preprocessor(X_val),  preprocessor(X_test)
```

Sebagai catatan, semua data yang telah dipreprocess tersebut akhirnya memusatkan nilainya ke 0, dan mengecilkan nilainya sehingga skalanya semua sama

Mencari mean squared error (rata-rata dari "error" y asli dan y prediksi). Dibawah adalah model sangat sederhana yaitu linear regression dari training set. Hasil yang diprint adalah mean squared error dari model tersebut pada training set.

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression

lm = LinearRegression().fit(X_train, y_train)
mse(lm.predict(X_train), y_train, squared=False), mse(lm.predict(X_val), y_val, squared=False) #mse dari training set dan validation set.
```

Error pada validation set lebih besar dari train set. Ini menunjukkan adanya overfitting. Berikut adalah model dengan K-nearest neighbor.

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors =20).fit(X_train, y_train)
mse(knn.predict(X_train), y_train, squared=False), mse(knn.predict(X_val), y_val, squared=False) #mse dari training set dan validation set.
```

Akhirnya, dengan K-Nearest neighbor, didapat dua mse yang tidak jauh berbeda sebelumnya. Ini menunjukkan performa model KNN lebih baik dalam mengeneralisir train set dan validation set.

Berikut adalah training dengan model random forest.

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(max_depth=10).fit(X_train, y_train)
mse(rfr.predict(X_train), y_train, squared=False), mse(rfr.predict(X_val), y_val, squared=False) #mse dari training set dan validation set.
```

Akhirnya, dengan RFR didapat error terendah, walau masih overfit. Kemudian, dicoba Gradient Boosting Regressor.

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(n_estimators=250).fit(X_train, y_train)
mse(gbr.predict(X_train), y_train, squared=False), mse(gbr.predict(X_val), y_val, squared=False) #mse dari training set dan validation set.
```
